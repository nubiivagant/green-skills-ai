```python
import numpy as np
import pandas as pd
```

```python
energy_data = pd.read_csv("energy_consumption.csv")
```

```python
energy_data
```

```python
energy_data.isna().sum()  # Check for missing values
```

```python
energy_data["Energy_Consumption(MWh)"].fillna(energy_data["Energy_Consumption(MWh)"].mean())
```

```python
energy_data["Energy_Consumption(MWh)"].fillna(energy_data["Energy_Consumption(MWh)"].mean(), inplace=True)
```

```python
energy_data.fillna(method="ffill")  # Forward fill missing values
```

```python
energy_data.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_scaler = MinMaxScaler()
energy_data[["Energy_Consumption(MWh)", "Cost(Million$)"]] = min_scaler.fit_transform(
	energy_data[["Energy_Consumption(MWh)", "Cost(Million$)"]]
)
```
